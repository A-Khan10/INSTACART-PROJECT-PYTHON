## Step 6 - Addressing Missing Values

In this course, we learned several ways to deal with missing.
One option is to remove or filter out missing data, Impute the value with the mean or median of the column, or create a new variable that acts as a flag based on the missing values.
However, the NaN in column "day_since_last_order" cannot be removed as it is the first order of a new customer, with no previous orders.

The second option is to impute values with mean or median, but this is not possible as it is based on the principle that if the value wasn't missing, it would be close to the mean or median value. In this case, the missing values could be zero or nothing, as the customer has no order before. Mean and median are not correct for imputed values, and zero as an imputed value may not be accurate due to the customer buying multiple orders on the same day.

The third option is to use NaN values as they are important as they indicate the customer's newness, as the missing value in this case means that the customer never had an order before.



### Addressing the missing values using an appropriate method.

We can create a new boolean column for "days_since_last_order" which will be True if NaN and False if not. 

```python
# creating new subset of "df_ords" as "df_ords_clean" and
# adding new column "new_customer" which will be True if NaN and False if not.
df_ords_clean = df_ords
df_ords_clean['new_customer'] = df_ords['days_since_last_order'].isnull() == True
df_ords_clean
```
